# **SIMPLE LINEAR REGRESSION**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
```

```python
df.head()
```

### **DATA ANALYSIS** 

```python
df.isnull().sum()
```

```python
df.shape
```

```python
print("No of unique values = " +str(df["Address"].nunique()))
df['Address'].value_counts()
```

From the above nunique we get that **we cant derive a relation between address and price as there is no pattern between them and every area is unique**. So we can now **drop** them.  

```python
dfc = df.drop(['Address'], axis = 1)
```

```python
dfc.head()
```

```python
dfc.describe()
```

```python
dfc.corr()
```

From the **correletion matrix** we get that **Avg. Area Income has the highest correlation with price.**

As we are performing **simple linear regression** we have to take **only one feature from the dataset**. So lets go with **Avg. Area Income as it has most correlation value with Price**.

```python
sns.pairplot(dfc, x_vars=['Avg. Area Income', 'Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population'], y_vars='Price', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
X = dfc['Avg. Area Income'].values.reshape(-1,1)
Y = dfc['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

### **MODEL TRAINING AND PREDICTION**

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
model = LinearRegression()  
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
```

```python
plt.scatter(X_test, Y_test,  color='gray')
plt.plot(X_test, Y_pred, color='red', linewidth=2)

plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
model.score(X_test, Y_test)
```

From the **accuracy score we can get that the accuracy score is not good which is only around 44%**. 

So we can conclude that **SIMPLE LINEAR REGRESSION is not a ideal model for data prediction in regression problem** as the problem **doesn't depend only on one features. It has multiple features.**
